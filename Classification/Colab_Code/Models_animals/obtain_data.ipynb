**Convert the data from drawings into features  by considering the offset between x and y coordinates of each drawing and save the data**



```
import numpy as np
import json
names = ["ant","bear","cat","butterfly","fish","bird","dog"]

for name in names:
  drawings = []
  ct = 0
  with open("/content/drive/My Drive/AML/RawData/"+name+".ndjson") as f:
    for line in f:
        ct += 1
        if(ct == 50000):
            break
        j_content = json.loads(line)
        drawing = j_content["drawing"]
        points = []
        points.append((0,0))
        for stroke in drawing:
            for i in range(1,len(stroke[0])):
                    offset_x = stroke[0][i]-stroke[0][i-1]
                    offset_y = stroke[1][i]-stroke[1][i-1]
                    points.append((offset_x,offset_y))
        drawings.append(points)
    
  drawings = np.array(drawings)
  print(drawings.shape)
  np.save("/content/drive/My Drive/AML/NPYData/"+name+"_points.npy",drawings)        
        
```
